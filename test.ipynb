```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt
data = iio.imread('bird.png')
plt.imshow(data)
plt.show()
```

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt
data = iio.imread('bird.png', mode='F')
plt.imshow(data)
plt.show()
```

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt
data = iio.imread('bird.png', mode='F').astype(np.uint8)
cl = data & 0xF0
iio.imsave('birdf0.png',cl)
tmp = iio.imread('birdf0.png')
plt.imshow(tmp)
plt.show()
```

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt
data = iio.imread('bird.png').astype(np.uint8)
cl = data & 0xc0
iio.imsave('birdc0.png',cl)
tmp = iio.imread('birdc0.png')
plt.imshow(tmp)
plt.show()
```

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt
data = iio.imread('bird.png').astype(np.uint8)
cl = data & 0x80
iio.imsave('bird80.png',cl)
tmp = iio.imread('bird80.png')
plt.imshow(tmp)
plt.show()
```

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt
data = iio.imread('bird.png')
bdata = (data[:,:,1] + data[:,:,2])
plt.imshow(bdata)
plt.show
```
